# Preprocess

## 1. select features

At first, we choose some features that may be important to new popularity, such as **title/topic/channel/week/time/length of text/number of image/sentiment/aurthor**.

However, in the training process, we find there are some features that are **useless**. Furthermore, these uesless features have bad impacts on test accuracy, so we remove the useless features.

We keep **topic/weekday/time/length of tex/sentiment** as our final features.  

## 2. encode
Fisrt, We use the **beautifulsoup** to find the html tag of the text, convert to lowercase and append all emoticons behind. 

Second, we encode them seperately:

    a. topic: we use CountVectorizer to encode topic.
    b. weekday: we use one hot encoding to encode weekday.
    c. time: we use one hot encoding to encode time.
    d. length of text: we doesn't encode textlen.
    e. sentiment: we use SentimentIntensityAnalyzer to vectorize the text.  

```python
#preprocessor for all text
def preprocessor(text):
    # remove HTML tags
    text = BeautifulSoup(text, 'html.parser').get_text()
    
    # regex for matching emoticons, keep emoticons, ex: :), :-P, :-D
    r = '(?::|;|=|X)(?:-)?(?:\)|\(|D|P)'
    emoticons = re.findall(r, text)
    text = re.sub(r, '', text)
    
    # convert to lowercase and append all emoticons behind (with space in between)
    # replace('-','') removes nose of emoticons
    text = re.sub('[\W]+', ' ', text.lower()) + ' ' + ' '.join(emoticons).replace('-','')
    
    return text
# print(preprocessor(df.loc[0,'Page content']))
```

```python
def preprocessor_title(X):
    ret=[]
    for text in X:
        titles = BeautifulSoup(text, 'html.parser')
        titles = titles.find_all('h1', class_='title')
        t=""
        for title in titles:
            t+=str(title.getText())
        r = '(?::|;|=|X)(?:-)?(?:\)|\(|D|P)'
        emoticons = re.findall(r, t)
        t = re.sub(r, '', t)
        t = re.sub('[\W]+', ' ', t.lower()) + ' ' + ' '.join(emoticons).replace('-','')
        ret.append(t)
    return ret
```

```python
def preprocessor_topic(X):
    ret=[]
    for text in X:
        topics = BeautifulSoup(text, 'html.parser')
        topics = topics.find_all(class_='article-topics')
        t=""
        for topic in topics:
            t+=str(topic.getText())

        r = '(?::|;|=|X)(?:-)?(?:\)|\(|D|P)'
        emoticons = re.findall(r, t)
        t = re.sub(r, '', t)
        t = re.sub('[\W]+', ' ', t.lower()) + ' ' + ' '.join(emoticons).replace('-','')
        t = t[7:] # remove " Topics"
        ret.append(t)
    
    return ret
```

```python
def preprocessor_channel(X):
    ret=[]
    for text in X:
        channel_flag = 0
        channels = BeautifulSoup(text, 'html.parser')
        channels = channels.find_all('article')
        for channel in channels:
            if 'data-channel' in channel.attrs:
                channel_flag = 1
                ret.append(channel['data-channel'])
        if channel_flag == 0:
            ret.append(" entertainment nfl sports  ")
    return ret
```

```python
def preprocessor_weekday(X):
    weekday = []
    for i in range(len(X)):
        if 'datetime="' in X[i]:
            split_data = X[i].split('datetime="')[1].split(",")[0]
            weekday.append(split_data)
        else:
            weekday.append("Mon")
    return weekday
```

```python
def preprocessor_time(X):
    ret=np.array([0, 0, 0, 0])
    z=0
    for text in X:
        one_doc = []
        times = BeautifulSoup(text, 'html.parser')
        times = times.find_all('time')
        
        
        sp = times[0].getText().split(' ')
        if(len(sp)!=1):
            date = sp[0].split('-')
            moment = sp[1].split(':')

            date = [int(d) for d in date]
            moment = [int(m) for m in moment]
            moment = moment[0]

            one_doc = np.hstack((np.array(date), np.array(moment)))
            ret = np.vstack((ret, one_doc))
        else:
            temp=np.array([2010, 9, 29, 8])
            ret = np.vstack((ret, temp))
    ret = np.delete(ret, 0, axis = 0)
    return ret
```

```python
def preprocessor_textlen(X):
    L=[]
    for text in X:
        inners = BeautifulSoup(text, 'html.parser')
        inners = inners.find_all('section', class_='article-content')
        #print(inners)
        L.append(len(inners[0].getText()))
    L = [int(l) for l in L]
    leng = len(L)
    L = np.array(L).reshape(leng, 1)
    return L
```

```python
stop = stopwords.words('english')
def tokenizer_stem_nostop(text):
    porter = PorterStemmer()
    return [porter.stem(w) for w in re.split('\s+', text.strip()) \
            if w not in stop and re.match('[a-zA-Z]+', w)]
```

```python
def vectorize_sentiment(X):
    ret=[]
    for text in X:
        # remove HTML tags
        sid = SentimentIntensityAnalyzer()
        text = BeautifulSoup(text, 'html.parser').get_text()
        # regex for matching emoticons, keep emoticons, ex: :), :-P, :-D
        r = '(?::|;|=|X)(?:-)?(?:\)|\(|D|P)'
        emoticons = re.findall(r, text)
        text = re.sub(r, '', text)
        # convert to lowercase and append all emoticons behind (with space in between)
        # replace('-','') removes nose of emoticons
        text = re.sub('[\W]+', ' ', text.lower()) + ' ' + ' '.join(emoticons).replace('-','')
        scores = sid.polarity_scores(text)
        temp=[]
        for key, value in scores.items():
            temp.append(value)        
        ret.append(temp[3])
    leng = len(ret)
    ret = np.array(ret).reshape(leng, 1)
    return ret
```

## MERGE FEATURE


After feature selection, we merge all the features together.

```python
TOPIC = preprocessor_topic(df['Page content'])    # (27643,  ) str, each is a str of all vacabulary in topic of a document
WEEK = preprocessor_weekday(df['Page content'])   # (27643,  ) str, each is one vacabulary from Mon to Sat
TIME = preprocessor_time(df['Page content'])      # (27643, 6) int, for each doc, (year, month, date, hour, minute, sec) 
TEXTLEN = preprocessor_textlen(df['Page content'])# (27643,  ) int, article-content length of each document
SENT = vectorize_sentiment(df['Page content'])
```

```python
WEEK_dum = pd.get_dummies(WEEK)
WEEK_dum = np.array(WEEK_dum)
```

```python
TOPIC_test = preprocessor_topic(df_test)    # (27643,  ) str, each is a str of all vacabulary in topic of a document
WEEK_test = preprocessor_weekday(df_test)   # (27643,  ) str, each is one vacabulary from Mon to Sat
TIME_test = preprocessor_time(df_test)      # (27643, 6) int, for each doc, (year, month, date, hour, minute, sec) 
TEXTLEN_test = preprocessor_textlen(df_test)# (27643,  ) int, article-content length of each document
SENT_test = vectorize_sentiment(df_test)
```

```python
WEEK_dum_test = pd.get_dummies(WEEK_test)
WEEK_dum_test = np.array(WEEK_dum_test)
```

```python
X_train = []
for i in range(len(TOPIC)):
    X_train.append(TOPIC[i])  
X_test = []
for i in range(len(TOPIC_test)):
    X_test.append(TOPIC_test[i])
```

```python
merge_data=np.hstack((np.array(X_train), np.array(X_test)))

vect = CountVectorizer(tokenizer=tokenizer_stem_nostop,
                      ).fit(merge_data)
X_train_vectorized = vect.transform(X_train)
X_test_vectorized = vect.transform(X_test)
Y_train=df['Popularity']
```

```python
dum = np.hstack((WEEK_dum, TIME))
dum = np.hstack((dum, TEXTLEN))
dum = np.hstack((dum, SENT))
X_train_vectorized = np.hstack((X_train_vectorized.toarray(), dum))


dum = np.hstack((WEEK_dum_test, TIME_test))
dum = np.hstack((dum, TEXTLEN_test))
dum = np.hstack((dum, SENT_test))
X_test_vectorized = np.hstack((X_test_vectorized.toarray(), dum))
```

# Start Training

After tryin plenty of models, we choose **LightGBM (LGBM)** as our model. LightGBM is a gradient boosting framework that uses tree based learning algorithms, and it has following advantages:

1. Faster training speed
2. Lower memory usage.
3. Capable of handling large-scale data

As for the parameter for LGBM, such as max_depth and n_estimetors, we use the knowledge learned in lab5 to find the best fit parameters, and to avoid overfit. Finally, we find out that the model has highest accuracy when max_depth=9 and n_estimators=5.

In order to get best test accuracy, We use for loop to select best training set and validation set.

```python
clf = LGBMClassifier(max_depth=9, n_estimators=50)
best_val_score=0
for i in range(500):
    x_train, X_val, y_train, y_val = train_test_split(X_train_vectorized, Y_train, test_size=0.2, shuffle=True)
    clf.fit(x_train, y_train)
    print("round:", i+1)
    train_score=clf.score(x_train, y_train)
    val_score=clf.score(X_val, y_val)
    
    print("train: ", train_score)
    print("valid: ", val_score)
    print("auc: ", roc_auc_score(y_val, clf.predict_proba(X_val)[:,1]))
    
    if(val_score>best_val_score):
        pred=clf.predict_proba(X_test_vectorized)[:,1]
        index=[]
        for i in range(len(pred)):
            index.append(27643+i)
        dict = {'Id': index, 'Popularity': pred}
        predict = pd.DataFrame(dict) 
        predict.to_csv('prediction.csv', index=False)
        best_val_score=val_score
        print("===============save model===============")
```
